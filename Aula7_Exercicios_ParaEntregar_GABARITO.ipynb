Exemplo extra - Dicionários

Lembretes importantes!!!

Listas → boas para armazenar coleções simples.

Dicionários → ótimos para estruturar dados com pares chave → valor.

Vamos estudar um exemplo utlizando o cálculo de média dos alunos:


```python

# Lista de alunos
alunos = ["Ana", "João", "Maria"]

# Matriz com notas
notas = [[7.5, 8.0, 9.0],[6.0, 5.5, 7.0],[9.0, 8.5, 10.0]]

# Criar lista de dicionários
dados_alunos = []

for i in range(len(alunos)):
    media = sum(notas[i]) / len(notas[i])
    aluno_dict = {
        "nome": alunos[i],
        "notas": notas[i],
        "media": media
    }
    dados_alunos.append(aluno_dict)
```

```python
print(dados_alunos)
```

```python
# Exibindo a média de todos os alunos
for aluno in dados_alunos:
    print(f"{aluno['nome']} - Notas: {aluno['notas']} - Média: {aluno['media']:.2f}")
```

```python
# Exibindo a média de um aluno específico diretamente, desde que pré-cadastrado
maria_data = None
nome = input("Digite o nome do aluno(a)")
nome = nome.lower().strip()
for aluno in dados_alunos:
    if aluno["nome"].lower() == nome:
        maria_data = aluno
        break

if maria_data:
    print(f"Média de {maria_data['nome']}: {maria_data['media']:.2f}, baseado nas notas: {maria_data['notas']}")
else:
    print("Aluno não encontrado.")
```
