EXTRA: Versão do código para checagem da idade:

```python
#Versão 4 (Considerando Dia, Mês e Ano) com 'continue':

import datetime

num_pessoas = int(input("Quantas pessoas você quer verificar a idade? "))

for i in range(num_pessoas):
  print(f"\nPessoa {i+1}:")
  ano_nascimento = int(input("Digite o ano de nascimento: "))

  hoje = datetime.date.today()
  ano_atual = hoje.year

  idade = ano_atual - ano_nascimento

  if idade > 18:
    print("Você é maior de idade.")
    continue # Pula para a próxima iteração se for maior de 18
  elif idade < 18:
    print("Você é menor de idade.")
    continue # Pula para a próxima iteração se for menor de 18
  else: # idade == 18
    mes_nascimento = int(input("Digite o mês de nascimento: "))
    dia_nascimento = int(input("Digite o dia de nascimento: "))

    mes_atual = hoje.month
    dia_atual = hoje.day

    if mes_atual < mes_nascimento:
       print("Você é menor de idade.")
    elif mes_atual == mes_nascimento:
        # Se o mês for o mesmo, verifica o dia
       if dia_atual < dia_nascimento:
           print("Você é menor de idade.")
       else:
           print("Você é maior de idade.")
    else:
        print("Você é maior de idade.")
```

```python
'''
Faça um programa em Python (utilize a estrutura for) que leia 10 valores inteiros e:
Encontre e mostre o maior valor
Encontre e mostre o menor valor
Calcule e mostre a média dos números lidos
'''


# 'inf' representa o infinito em Python. Quando você usa float('inf') ou float('-inf'), está criando um valor que é maior ou menor do que qualquer número real.
#Isso é útil para inicializar uma variável que você usará para encontrar o maior ou menor valor em uma sequência, garantindo que
#o primeiro número da sequência seja sempre maior ou menor do que o valor inicial e se torne o primeiro "menor valor" ou "maior valor" encontrado.
maior_valor = float('-inf')
menor_valor = float('inf')
soma = 0

for i in range(10):
  valor = int(input(f"Digite o {i+1}º valor inteiro: "))
  soma += valor #soma = soma + valor

  if valor > maior_valor:
    maior_valor = valor
  if valor < menor_valor:
    menor_valor = valor

media = soma / 10

print(f"\nO maior valor digitado é: {maior_valor}")
print(f"O menor valor digitado é: {menor_valor}")
print(f"A média dos valores digitados é: {media}")
```

```python
'''
Faça um programa em linguagem Python que recebe
a temperatura de n clientes e imprima a mensagem
se a temperatura está normal (menor que 37,2 C),
se está em estado febril (37,3 C a 38 C ), com
febre (38 C a 39 C) ou com febre alta(acima 39 C).
No final mostre a quantidade de pessoas analisadas
e a média de temperatura.
'''

num_clientes = int(input("Quantos clientes você quer verificar a temperatura? "))
soma_temperaturas = 0
pessoas_analisadas = 0

for i in range(num_clientes):
  temperatura = float(input(f"Digite a temperatura do cliente {i+1}: "))
  soma_temperaturas += temperatura
  pessoas_analisadas += 1

  if temperatura < 37.2:
    print("Temperatura normal.")
  elif 37.3 <= temperatura <= 38:
    print("Estado febril.")
  elif 38 < temperatura <= 39:
    print("Com febre.")
  else:
    print("Com febre alta.")

if pessoas_analisadas > 0:
  media_temperatura = soma_temperaturas / pessoas_analisadas
  print(f"\nQuantidade de pessoas analisadas: {pessoas_analisadas}")
  print(f"Média de temperatura: {media_temperatura:.2f} C")
else:
  print("\nNenhuma pessoa foi analisada.")
```

```python
'''
Faça um programa que leia um nome de usuário e a
sua senha e não aceite a senha igual ao nome do
usuário, mostrando uma mensagem de erro e voltando
a pedir as informações.
'''

while True:
  nome_usuario = input("Digite o nome de usuário: ")
  senha = input("Digite a senha: ")

  if senha == nome_usuario:
    print("Erro: A senha não pode ser igual ao nome de usuário. Tente novamente.")
  else:
    print("Nome de usuário e senha aceitos.")
    break

print("Fim do programa.")
```

```python

'''
Faça um programa que solicite ao usuário um número que ele queira
treinar a tabuada.
Você irá solicitar ao mesmo a resposta do cálculo do número
informado multiplicado por 1, 2 até 10.
A cada resposta você deverá validar e imprimir :”CORRETO” ou
“QUE PENA, VOCÊ ERROU, O VALOR CORRETO É X “, no lugar de ”X“
coloque o valor correto.
Ao final imprima “Total de acertos: y” e “Total de erros z”,
onde “y“ deverá ser o total de acertos e “z“ o total de erros.
'''

numero_tabuada = int(input("Digite o número que você quer treinar a tabuada: "))

acertos = 0
erros = 0

for i in range(1, 11):
  resposta_correta = numero_tabuada * i
  resposta_usuario = int(input(f"{numero_tabuada} x {i} = "))

  if resposta_usuario == resposta_correta:
    print("CORRETO")
    acertos += 1
  else:
    print(f"QUE PENA, VOCÊ ERROU, O VALOR CORRETO É {resposta_correta}")
    erros += 1

print(f"\nTotal de acertos: {acertos}")
print(f"Total de erros: {erros}")
```
