EXERCÍCIO 1

```python
'''
Escreva um programa que solicite vários números ao usuário,
sendo um de cada vez, possibilitando encerrar a entrada de dados
informando zero.
Adicione os números informados em uma lista e, ao final do programa,
imprima a soma de todos os números, a multiplicação de todos os
números, o maior e o menor número informado.
'''

numeros = []
while True:
    num_str = input("Digite um número inteiro (ou 0 para sair): ")
    try:
        num = int(num_str)
        if num == 0:
            break
        numeros.append(num)
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")

if numeros:
    soma = sum(numeros)

    produto = 1
    for n in numeros:
        produto *= n

    maior = max(numeros)
    menor = min(numeros)

    print(f"\nNúmeros informados: {numeros}")
    print(f"Soma dos números: {soma}")
    print(f"Multiplicação dos números: {produto}")
    print(f"Maior número: {maior}")
    print(f"Menor número: {menor}")
else:
    print("Nenhum número foi informado.")
```

EXERCÍCIO 2

```python
'''
Escreva um programa que solicite várias palavras ao usuário,
sendo uma de cada vez, possibilitando encerrar a entrada de dados
informando zero.
Adicione as palavras informadas em uma lista. Após o usuário informar
“0” (zero), solicite a palavra que ele deseja contar.
Desta forma, o programa deverá contar as ocorrências daquela palavra. Ao final imprima o resultado.
'''

palavras = []
while True:
    palavra = input("Digite uma palavra (ou 0 para sair): ")
    if palavra == '0':
        break
    palavras.append(palavra)

if palavras:
    palavra_contar = input("Digite a palavra que deseja contar: ")
    contagem = palavras.count(palavra_contar)
    print(f"A palavra '{palavra_contar}' aparece {contagem} vezes na lista.")
else:
    print("Nenhuma palavra foi informada.")
```

```python
# Exemplo para demonstrar o uso de set

# Uma lista com elementos duplicados
my_list = [1, 2, 2, 3, 4, 4, 5]
print(f"Lista original: {my_list}")

# Converte a lista para um conjunto para remover duplicados
my_set = set(my_list)
print(f"Conjunto com elementos únicos: {my_set}")

# Converte o conjunto de volta para uma lista (opcional)
unique_list = list(my_set)
print(f"Lista com elementos únicos: {unique_list}")
```

```python
# Exemplo para demonstrar a diferença entre listas e conjuntos

# Uma lista pode conter elementos duplicados e mantém a ordem
my_list = [1, 2, 2, 3, 4, 4, 5]
print(f"Lista: {my_list}")
print(f"Tipo de lista: {type(my_list)}")

# Um conjunto contém apenas elementos únicos e não mantém a ordem
my_set = {1, 2, 2, 3, 4, 4, 5}
print(f"Conjunto: {my_set}")
print(f"Tipo de conjunto: {type(my_set)}")
```

```python
# Criando dois conjuntos
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

print(f"Conjunto 1: {set1}")
print(f"Conjunto 2: {set2}")

# União de conjuntos
union_set = set1.union(set2)
print(f"União dos Conjuntos 1 e 2: {union_set}")

# Interseção de conjuntos
intersection_set = set1.intersection(set2)
print(f"Interseção dos Conjuntos 1 e 2: {intersection_set}")

# Diferença de conjuntos (elementos em set1, mas não em set2)
difference_set1 = set1.difference(set2)
print(f"Diferença dos Conjuntos 1 e 2: {difference_set1}")

# Diferença de conjuntos (elementos em set2, mas não em set1)
difference_set2 = set2.difference(set1)
print(f"Diferença dos Conjuntos 2 e 1: {difference_set2}")

# Diferença simétrica (elementos em set1 ou set2, mas não em ambos)
symmetric_difference_set = set1.symmetric_difference(set2)
print(f"Diferença Simétrica dos Conjuntos 1 e 2: {symmetric_difference_set}")

# Verificar se um elemento está em um conjunto
print(f"O 3 está no Conjunto 1? {3 in set1}")
print(f"O 9 está no Conjunto 1? {9 in set1}")
```

```python
'''
Escreva um programa que solicite vários números ao usuário, sendo um de cada vez,
possibilitando encerrar a entrada de dados informando zero.
Adicione os números em uma lista, crie uma outra lista contendo os números sem
repetição e uma outra contendo os números que se repetem na primeira lista.
'''
numeros = []
while True:
    num_str = input("Digite um número inteiro (ou 0 para sair): ")
    try:
        num = int(num_str)
        if num == 0:
            break
        numeros.append(num)
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")

if numeros:
    numeros_unicos = list(set(numeros)) # Cria uma lista com os números únicos usando um conjunto (set remove elementos duplicados).

    numeros_repetidos = []
    vistos = set()
    for num in numeros:
        if num in vistos:
            numeros_repetidos.append(num)
        else:
            vistos.add(num)
        print(numeros_repetidos)
        print(vistos)

    print(f"\nNúmeros informados: {numeros}")
    print(f"Números sem repetição: {numeros_unicos}")
    print(f"Números que se repetem: {numeros_repetidos}")
else:
    print("Nenhum número foi informado.")
```

DESAFIO

```python
'''
Escreva um programa que solicite vários números ao usuário, sendo um de cada vez, possibilitando encerrar a entrada de dados informando zero.
Armazene os números em uma lista, depois percorra esta lista alimentando outras duas listas, uma com números pares e outra com números ímpares.
Ordene e imprima os números em ordem crescente e some os valores e imprima o resultado.
'''

numeros = []
while True:
    num_str = input("Digite um número inteiro (ou 0 para sair): ")
    try:
        num = int(num_str)
        if num == 0:
            break
        numeros.append(num)
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")

if numeros:
    pares = []
    impares = []

    for n in numeros:
        if n % 2 == 0:
            pares.append(n)
        else:
            impares.append(n)

    pares.sort()
    impares.sort()

    soma_pares = sum(pares)
    soma_impares = sum(impares)

    print(f"\nNúmeros informados: {numeros}")
    print(f"Números pares (ordenados): {pares}")
    print(f"Números ímpares (ordenados): {impares}")
    print(f"Soma dos números pares: {soma_pares}")
    print(f"Soma dos números ímpares: {soma_impares}")
else:
    print("Nenhum número foi informado.")
```
