```python
palavras = []
while True:
  p = input("Digite uma palavra: ").lower().strip()
  if p == "SAIR".lower().strip():
    break
  palavras.append(p)
print(f"As palavras digitadas foram: {palavras}")
conta = input("Diga a palavra que deseja contar: ").lower().strip()
quantidade = palavras.count(conta)
print(f"A quantidade de palavras '{conta}' é igual a {quantidade}")
```

```python
numeros = []
while True:
  num_str = input("Digite um numero inteiro(0 para sair): ")
  try:
    num = int(num_str)
    if num == 0:
      break
    numeros.append(num)
  except ValueError:
    print("Entrada inválida. Por favor, digite um número inteiro.")
if numeros:
  numeros_unicos = list(set(numeros))

  numeros_repetidos = []
  vistos = set()
  for num in numeros:
    if num in vistos:
      numeros_repetidos.append(num)
    else:
      vistos.add(num)
  print(f"Números informados: {numeros}")
  print(f"Números sem repetição: {numeros_unicos}")
  print(f"Números repetidos: {numeros_repetidos}")
else:
  print("Nenhum número foi informado.")
```
