```python
p1 = float(input("Digite a nota da prova número 1: "))
p2 = float(input("Digite a nota da prova número 2: "))
p3 = float(input("Digite a nota da prova número 3: "))

media = (p1 + p2 + p3) / 3

print("Sua média foi de : ", media)
```

```python
catetoA = 4
catetoB = 3

hipotenusa = ((catetoA ** 2) + (catetoB ** 2)) ** 0.5

print("A hipotenusa é igual a: ", hipotenusa)
```

```python
salario = float(input("Digite seu salário atual: "))
percentual = float(input("Digite o percentual de aumento: "))
aumento = salario * (percentual / 100)
salarioAtualizado = salario + aumento

print("Seu salário atualizado é de ", salarioAtualizado)
```

```python
PI = 3.14
raio = float(input("Digite o raio da lata: "))
altura = float(input("Digite a altura da lata: "))

volume = PI * raio ** 2 * altura

print("O volume da sua lata é de ", volume)
```
