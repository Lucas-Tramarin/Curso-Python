```python
produtos = {
    "Mouse": 98.75,
    "Teclado": 125.65,
    "Monitor": 134.78,
    "Gabinte": 170.00,
    "HD Externo": 510.50,
    "Headset": 125.45
}
while True:
  produto = input("Informe o produto a pesquisar o preço ou fim para sair: ")
  if produto == "fim":
    break
  if produto in produtos:
    print(f"Produto {produto} custa {produtos[produto]}")
  else:
    print(f"Produto {produto} não encontrado")
```

```python
aluno = {}
aluno["nome"] = input("Digite o nome do aluno: ")
aluno["media"] = float(input(f"Digite a média do aluno {aluno["nome"]}: "))
if aluno["media"] >= 7:
  aluno["situacao"] = "Aprovado"
elif 5 <= aluno["media"] < 7:
  aluno["situacao"] = "Recuperação"
else:
  aluno["situacao"] = "Reprovado"
print("=="*30)
for k, v in aluno.items():
  print(f"{k} = {v}")
```

```python
semana = {}

semana['segunda'] = ['python', 'java', 'ioga']
semana['terca'] = []
semana['quarta'] = ['luta livre']
semana['quinta'] = []
semana['sexta'] = []
semana['sabado'] = []
semana['domingo'] = []

while True:
  print("\n--- Programação Atual ---")
  for dia, aulas in semana.items():
    print(f"{dia.capitalize()}: {aulas}")

  dia_para_modificar = input("\nDigite o dia da semana que você quer modificar (segunda, terca, quarta, quinta, sexta, sabado, domingo), ou 'fim' para terminar: ").lower()

  if dia_para_modificar == 'fim':
    break

  if dia_para_modificar in semana:
    aulas_input = input(f"Digite as aulas para {dia_para_modificar}, separadas por vírgulas: ")

    aulas_lista = [aula.strip() for aula in aulas_input.split(',')]

    semana[dia_para_modificar] = aulas_lista
    print(f"\nHorário para {dia_para_modificar} atualizado.")
  else:
    print(f"\nDia '{dia_para_modificar}' inválido. Por favor, digite um dia válido ou 'fim' para terminar.")

print("\n--- Programação Final ---")
for dia, aulas in semana.items():
  print(f"{dia.capitalize()}: {aulas}")

print("\nPrograma encerrado.")
```

```python
filmes = {}

filmes['Star Wars: A New Hope'] = {'vilão': 'Darth Vader', 'ano': 1977}
filmes['The Dark Knight'] = {'vilão': 'Joker', 'ano': 2008}
filmes['The Matrix'] = {'vilão': 'Agent Smith', 'ano': 1999}

print("Informações sobre Star Wars: A New Hope:")
print(f"Vilão: {filmes['Star Wars: A New Hope']['vilão']}")
print(f"Ano de lançamento: {filmes['Star Wars: A New Hope']['ano']}")

print("\nInformações sobre The Dark Knight:")
print(f"Vilão: {filmes['The Dark Knight']['vilão']}")
print(f"Ano de lançamento: {filmes['The Dark Knight']['ano']}")

print("\nInformações sobre The Matrix:")
print(f"Vilão: {filmes['The Matrix']['vilão']}")
print(f"Ano de lançamento: {filmes['The Matrix']['ano']}")
```

```python
print("Iterando sobre os filmes: ")
for filme, info in filmes.items():
  print(f"Filmes: {filme}")
  print(f"Vilão: {info['vilão']}")
  print(f"Ano: {info['ano']}")
  print("=" * 30)
```
