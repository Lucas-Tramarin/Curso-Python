```python
produtos = {
    "Mouse": 98.75,
    "Teclado": 125.65,
    "Monitor": 134.78,
    "Gabinte": 170.00,
    "HD Externo": 510.50,
    "Headset": 125.45
}
while True:
  produto = input("Informe o produto a pesquisar o preço ou fim para sair: ")
  if produto == "fim":
    break
  if produto in produtos:
    print(f"Produto {produto} custa {produtos[produto]}")
  else:
    print(f"Produto {produto} não encontrado")
```

```python
aluno = {}
aluno["nome"] = input("Digite o nome do aluno: ")
aluno["media"] = float(input(f"Digite a média do aluno {aluno["nome"]}: "))
if aluno["media"] >= 7:
  aluno["situacao"] = "Aprovado"
elif 5 <= aluno["media"] < 7:
  aluno["situacao"] = "Recuperação"
else:
  aluno["situacao"] = "Reprovado"
print("=="*30)
for k, v in aluno.items():
  print(f"{k} = {v}")
```
