```python
'''O método de swap método funciona para qualquer tipo de dado e linguagem de programação.
Declare uma variável temporária (aux): Copie o valor de uma das variáveis (por exemplo, x) para a variável aux.
Copie o valor da segunda para a primeira: Atribua o valor de y a x.
Atribua o valor temporário: Copie o valor de aux para y.
Exemplo (pseudocódigo):
aux = a
a = b
b = aux'''

#Exemplo com valores:

a = input("Digite o valor de a: ")
b = input("Digite o valor de b: ")
print("A variável a é igual a: ", a)
print("A variável b é igual b: ", b)
```

```python
aux = a
a = b
b = aux
print("A variável a é igual a:", a)
print("A variável b é igual a:", b)
```

```python
'''O Python permite a troca de valores em uma única linha de código, simplificando o processo. Use atribuição múltipla:
Atribua diretamente os valores de b para a e de a para b. Exemplo (Python): a, b = b, a'''
a = 17
b = 4
a, b = b, a
print("A variável a é igual a:", a)
print("A variável b é igual a:", b)
```

```python
#Podemos, assim, atribuir valor a muitas variáveis por vez:
var1,var2,var3 = "Oi","Tudo bem?","Como vai"
print(var1)
print(var2)
print(var3)
```

```python
'''Em Python você pode atribuir valores a quantas variáveis quiser em uma única linha, separando as variáveis e os valores por vírgulas,como no exemplo anterior.
Não há um limite rígido imposto pela linguagem Python em si para a quantidade de variáveis que você pode atribuir em uma única linha dessa forma.
No entanto, por questões de legibilidade e boas práticas de código, é recomendável não colocar um número excessivo de variáveis em uma única linha,
pois isso pode tornar o código difícil de ler e entender. O ideal é usar essa funcionalidade para um número razoável de variáveis, onde isso melhora a clareza do código.'''
```

```python
#Dadas as notas das provas P1, P2 e P3, calcular a média (aritmética) das notas do aluno:

P1 = float(input("Digite a nota da prova 1: "))
P2 = float(input("Digite a nota da prova 2: "))
P3 = float(input("Digite a nota da prova 3: "))

media = (P1 + P2 + P3) / 3
media_arred = round(media,2)

print(media)
print(f"A média das notas é: {media_arred}")
```

```python
#Escreva um código que calcule a hipotenusa de um triângulo retângulo, cujos catetos são a = 4 e b = 3.
cateto_a = 4
cateto_b = 3
c = (cateto_a ** 2 + cateto_b ** 2) ** (1/2)
print(f"A hipotenusa do triângulo é: {c}")
```

```python
#Versão do exercício acima usando o método sqrt da biblioteca math:
import math

cateto_a = 4
cateto_b = 3
c = math.sqrt(cateto_a ** 2 + cateto_b ** 2)
print(f"A hipotenusa do triângulo é: {c}")
```

```python
'''Solicite ao usuário o valor do salário atual (numérico com
decimais), em seguida, solicite o percentual de aumento
 (numérico com decimais) e imprima o valor do salário atualizado'''

salario = float(input("Digite o valor do salário atual: "))
percentual_aumento = float(input("Digite o percentual de aumento: "))
aumento = salario * (percentual_aumento / 100)
novo_salario = salario + aumento
print(f"O valor do salário atualizado é: {novo_salario}")
```

```python
'''Crie um programa e declare uma constante PI
(use 2 casas após a vírgula). Dados o raio e a
altura, calcular e apresentar o valor do volume de
 uma lata de óleo,
utilizando a fórmula: volume = PI * r 2 * altura'''

PI = 3.14
raio = float(input("Digite o raio da lata de óleo: "))
altura = float(input("Digite a altura da lata de óleo: "))
volume = PI * raio ** 2 * altura
print(f"O volume da lata de óleo é: {volume}")
```
