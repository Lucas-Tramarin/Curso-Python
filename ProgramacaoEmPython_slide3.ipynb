```python
#Versão 3 (Considerando Dia, Mês e Ano):

import datetime

ano_nascimento = int(input("Digite o ano de nascimento: "))
mes_nascimento = int(input("Digite o mês de nascimento: "))
dia_nascimento = int(input("Digite o dia de nascimento: "))


hoje = datetime.date.today()
ano_atual = hoje.year
mes_atual = hoje.month
dia_atual = hoje.day


idade = ano_atual - ano_nascimento
if mes_atual < mes_nascimento:
   idade -= 1 # mesma coisa que "idade = idade - 1"
elif mes_atual == mes_nascimento:
    # Se o mês for o mesmo, verifica o dia
   if dia_atual < dia_nascimento:
       idade -= 1

print("A sua idade é:", idade)

if idade >= 18:
  print("Você é maior de idade.")
else:
  print("Você é menor de idade.")
```

```python
'''
Elabore um algoritmo que dada a idade de um nadador classifique-o em uma das
seguintes categorias:
Infantil A = 5 a 7 anos
Infantil B = 8 a 11 anos
Juvenil A = 12 a 13 anos
Juvenil B = 14 a 17 anos
Adultos = Maiores de 18 anos
'''

idade_nadador = int(input("Digite a idade do nadador: "))

if 5 <= idade_nadador <= 7:#if idade_nadador>=5 and idade_nadador<=7
  print("Categoria: Infantil A")
elif 8 <= idade_nadador <= 11:
  print("Categoria: Infantil B")
elif 12 <= idade_nadador <= 13:
  print("Categoria: Juvenil A")
elif 14 <= idade_nadador <= 17:
  print("Categoria: Juvenil B")
elif idade_nadador >= 18:
  print("Categoria: Adultos")
else:
  print("Idade não se encaixa em nenhuma categoria definida.")
```

```python
'''
Faça um programa que solicite dois números ao usuário (com decimais).
Em seguida solicite que o usuário informe o resultado das quatro
operações matemáticas ( subtração, multiplicação e divisão)
'''

num1 = float(input("Digite o primeiro número com decimal: "))
num2 = float(input("Digite o segundo número com decimal: "))

# Solicita os resultados das operações ao usuário
resultado_soma_usuario = float(input(f"Qual o resultado da soma de {num1} + {num2}? "))
resultado_subtracao_usuario = float(input(f"Qual o resultado da subtração de {num1} - {num2}? "))
resultado_multiplicacao_usuario = float(input(f"Qual o resultado da multiplicação de {num1} * {num2}? "))

# Verifica se o segundo número não é zero antes de solicitar a divisão
if num2 != 0:
  resultado_divisao_usuario = float(input(f"Qual o resultado da divisão de {num1} / {num2}? "))
else:
  print("Não é possível realizar a divisão por zero.")
  resultado_divisao_usuario = None # Define como None se a divisão não for possível

print()

# lógica para verificar se as respostas do usuário estão corretas

if resultado_soma_usuario == (num1 + num2):
  print("A soma está correta!\n")
else:
  print("A soma está incorreta.\n")

if resultado_subtracao_usuario == (num1 - num2):
  print("A subtração está correta!\n")
else:
  print("A subtração está incorreta.\n")

if resultado_multiplicacao_usuario == (num1 * num2):
  print("A multiplicação está correta!\n")
else:
  print("A multiplicação está incorreta.\n")

if resultado_divisao_usuario == (num1 / num2):
  print("A divisão está correta!\n")
else:
  print("A divisão está incorreta.\n")
```
