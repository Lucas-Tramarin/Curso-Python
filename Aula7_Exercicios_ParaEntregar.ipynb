Exercício para entrega - Dicionários

Lembretes importantes!!!

Listas → boas para armazenar coleções simples.

Dicionários → ótimos para estruturar dados com pares chave → valor.

```python
# Exercício 1:

# Você possui duas listas: `alunos`, contendo os nomes dos estudantes,
# e `notas`, uma matriz onde cada sublista corresponde às notas de um aluno
# na mesma ordem da lista `alunos`.
# Seu objetivo é criar uma nova estrutura de dados, uma lista de dicionários
# chamada `dados_alunos`. Cada dicionário nesta lista deve representar um aluno
# e conter as seguintes chaves e valores:
# - `nome`: o nome do aluno (string)
# - `notas`: a lista de notas do aluno (lista de floats)
# - `media`: a média aritmética das notas do aluno (float)
#
# Você deve iterar sobre os dados existentes, calcular a média para cada aluno
# e popular a lista `dados_alunos` com os dicionários correspondentes.
```

```python
alunos = ["joao","maria","jose"]
notas = [
    [10.0, 8.5, 9.0],
    [3.0, 5.0, 2.0],
    [6.0, 7.0, 5.5]
]

dados_alunos = []

for i in range(len(alunos)):
  media = sum(notas[i]) / len(notas[i])

  aluno_dict = {
    "nome": alunos[i],
    "notas": notas[i],
    "media": media
  }

  dados_alunos.append(aluno_dict)

for aluno in dados_alunos:
  print(aluno)
```

```python
# Exercício 2:

# Assumindo que você já possui a lista `dados_alunos` (uma lista de dicionários,
# onde cada dicionário representa um aluno com chaves como 'nome', 'notas' e 'media'),
# escreva um código para iterar sobre esta lista e exibir o nome do aluno,
# suas notas e a média calculada, formatando a média com duas casas decimais.
# A saída deve ser clara e legível para cada aluno.
```

```python
for aluno in dados_alunos:
  nome = aluno["nome"]
  notas = aluno["notas"]
  media = aluno["media"]

  print(f"Aluno: {nome}")
  print(f"Notas: {notas}")
  print(f"Média: {media:.2f}")
  print("-" * 30)
```

```python
# Exercício 3:

# Com base na lista `dados_alunos` criada anteriormente, implemente um código
# que permita ao usuário buscar por um aluno específico. O programa deve solicitar
# o nome do aluno via entrada do usuário (`input`). Se o aluno for encontrado na lista,
# exiba seu nome, notas e a média formatada com duas casas decimais. Caso contrário,
# informe que o aluno não foi encontrado.
```

```python
busca = input("Digite o nome do aluno: ").strip()

encontrado = False

for aluno in dados_alunos:
  if aluno["nome"].lower() == busca.lower():
    print(f"Aluno encontrado!")
    print(f"Nome: {aluno['nome']}")
    print(f"Notas: {aluno['notas']}")
    print(f"Média: {aluno['media']:.2f}")
    encontrado = True
    break

if not encontrado:
  print("Aluno não encontrado.")
```
